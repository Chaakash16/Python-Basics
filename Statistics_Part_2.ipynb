**Statistics Part 2**

1. **What is hypothesis testing in statistics?**  
   Hypothesis testing is a statistical method to test assumptions about a population parameter using sample data.  

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**  
   The null hypothesis (H₀) assumes no effect or difference, while the alternative hypothesis (H₁) suggests a significant effect or difference.  

3. **What is the significance level in hypothesis testing, and why is it important?**  
   The significance level (α) is the probability of rejecting a true null hypothesis, typically set at 0.05 (5%).  

4. **What does a P-value represent in hypothesis testing?**  
   The P-value represents the probability of obtaining the observed results if the null hypothesis is true.  

5. **How do you interpret the P-value in hypothesis testing?**  
   A small P-value (≤ α) suggests rejecting H₀; a large P-value (> α) suggests failing to reject H₀.  

6. **What are Type 1 and Type 2 errors in hypothesis testing?**  
   Type 1 error (α) is rejecting a true H₀; Type 2 error (β) is failing to reject a false H₀.  

7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**  
   A one-tailed test checks for a directional effect, while a two-tailed test checks for any difference.  

8. **What is the Z-test, and when is it used in hypothesis testing?**  
   A Z-test is used to compare population means when the sample size is large (n ≥ 30) and variance is known.  

9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**  
   The Z-score measures how many standard deviations a value is from the mean; Z = (X - μ) / σ.  

10. **What is the T-distribution, and when should it be used instead of the normal distribution?**  
    The T-distribution is used for small sample sizes (n < 30) when the population variance is unknown.  

11. **What is the difference between a Z-test and a T-test?**  
    A Z-test is for large samples with known variance, while a T-test is for small samples with unknown variance.  

12. **What is the T-test, and how is it used in hypothesis testing?**  
    A T-test compares the means of two groups to determine if they are significantly different.  

13. **What is the relationship between the Z-test and T-test in hypothesis testing?**  
    Both compare means, but a T-test adjusts for smaller samples and unknown variance.  

14. **What is a confidence interval, and how is it used to interpret statistical results?**  
    A confidence interval is a range of values estimating a population parameter with a given confidence level (e.g., 95%).  

15. **What is the margin of error, and how does it affect the confidence interval?**  
    The margin of error defines the range within which the true population parameter is expected to lie, affecting the precision of the confidence interval.  

16. **How is Bayes' Theorem used in statistics, and what is its significance?**  
    Bayes' Theorem updates probabilities based on new evidence and is crucial in Bayesian statistics.  

17. **What is the Chi-square distribution, and when is it used?**  
    The Chi-square distribution is used for categorical data analysis and hypothesis testing for variance and independence.  

18. **What is the Chi-square goodness of fit test, and how is it applied?**  
    The Chi-square goodness of fit test checks if observed data matches an expected distribution.  

19. **What is the F-distribution, and when is it used in hypothesis testing?**  
    The F-distribution is used in variance comparison tests like ANOVA and F-tests.  

20. **What is an ANOVA test, and what are its assumptions?**  
    ANOVA analyzes variance among multiple groups and assumes normality, independence, and equal variance.  

21. **What are the different types of ANOVA tests?**  
    One-way ANOVA (one factor), Two-way ANOVA (two factors), and Repeated Measures ANOVA.  

22. **What is the F-test, and how does it relate to hypothesis testing?**  
    The F-test compares variances of two or more groups to determine if they are significantly different.

Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
import scipy.stats as stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    print(f"Sample Mean: {sample_mean}")
    print(f"Population Mean: {pop_mean}")
    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")

    if p_value < alpha:
        print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)
population_mean = 105
population_std = 15

z_test(sample_data, population_mean, population_std)
```

Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=100)
population_mean = 52
population_std = 10

z_stat = (np.mean(sample) - population_mean) / (population_std / np.sqrt(len(sample)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)

if abs(z_stat) > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Standard Normal Distribution')
plt.axvline(-z_critical, color='red', linestyle='dashed', label='Rejection Region')
plt.axvline(z_critical, color='red', linestyle='dashed')
plt.axvline(z_stat, color='blue', linestyle='solid', label='Test Statistic')
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3)
plt.legend()
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.title('Two-Tailed Z-test Decision Region')
plt.show()
```

Q6. Write a Python program to perform an independent T-test and interpret the results.

Q7. Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def perform_paired_ttest(before, after, alpha=0.05):
    t_stat, p_value = stats.ttest_rel(before, after)
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between paired samples.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between paired samples.")

    return t_stat, p_value

np.random.seed(42)
before = np.random.normal(loc=50, scale=10, size=100)
after = before + np.random.normal(loc=2, scale=5, size=100)

perform_paired_ttest(before, after)

plt.figure(figsize=(8, 5))
plt.scatter(before, after, alpha=0.7, label='Paired Data')
plt.plot([min(before), max(before)], [min(before), max(before)], color='red', linestyle='dashed', label='No Change Line')
plt.xlabel('Before')
plt.ylabel('After')
plt.title('Paired Sample T-test Comparison')
plt.legend()
plt.show()
```
