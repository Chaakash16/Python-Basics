##**Theory Questions**

Q1. What is the difference between a function and a method in Python?

Ans. A **function** is a standalone block of code (e.g., `def greet():`), while a **method** is a function bound to an object or class (e.g., `class MyClass: def greet(self):`).

Q2. Explain the concept of function arguments and parameters in Python.

Ans. In Python, **parameters** are the variables in a function definition (e.g., `def add(a, b):`), and **arguments** are the values passed when calling the function (e.g., `add(3, 4)`).

Q3. What are the different ways to define and call a function in Python?

Ans. Functions in Python can be defined using def (e.g., def greet():) and called with parentheses (e.g., greet()), or using lambda expressions for anonymous functions (e.g., square = lambda x: x**2; square(4)).

Q4. What is the purpose of the `return` statement in a Python function?

Ans. The `return` statement in a Python function is used to send a result back to the caller (e.g., `def add(a, b): return a + b`).

Q5. What are iterators in Python and how do they differ from iterables?

Ans. An **iterator** in Python is an object that implements `__iter__()` and `__next__()` methods, allowing it to be traversed (e.g., `it = iter([1, 2, 3])`), while an **iterable** is any object that can return an iterator (e.g., a list like `[1, 2, 3]`).

Q6. Explain the concept of generators in Python and how they are defined.

Ans. Generators in Python are functions that yield values lazily using the `yield` keyword (e.g., `def count_up_to(n): while n > 0: yield n; n -= 1`).

Q7. What are the advantages of using generators over regular functions?

Ans. Generators are memory-efficient because they yield values one at a time instead of returning all values at once (e.g., def gen(): yield 1; yield 2), unlike regular functions that return a complete list.

Q8.  What is a lambda function in Python and when is it typically used?

Ans. A lambda function in Python is a small, anonymous function defined using the lambda keyword (e.g., add = lambda x, y: x + y), typically used for short, throwaway functions where defining a full function is unnecessary.

Q9. Explain the purpose and usage of the `map()` function in Python.

Ans. The map() function in Python applies a given function to all items in an iterable and returns an iterator (e.g., map(lambda x: x**2, [1, 2, 3]) results in [1, 4, 9]).

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans. In Python:
*   map() applies a function to all items in an iterable and returns an iterator (e.g., map(lambda x: x+1, [1, 2, 3]) → [2, 3, 4]).
*   reduce() (from functools) applies a function cumulatively to the items of an iterable, reducing it to a single value (e.g., reduce(lambda x, y: x + y, [1, 2, 3]) → 6).
*   filter() filters elements from an iterable based on a condition provided by a function, returning an iterator with the elements that satisfy the condition (e.g., filter(lambda x: x % 2 == 0, [1, 2, 3]) → [2]).

Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
    total = 0

    for num in numbers:

        if num % 2 == 0:
            total += num

    return total

 #Example

numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print(result)  # Output: 12 (2 + 4 + 6)
```

Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

#Example

for num in range(1, 201):
    if is_prime(num):
        print(num)
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.prev, self.next = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.terms:
            raise StopIteration
        if self.current == 0:
            self.current += 1
            return 0
        elif self.current == 1:
            self.current += 1
            return 1
        else:
            fib_number = self.prev + self.next
            self.prev, self.next = self.next, fib_number
            self.current += 1
            return fib_number

#Example

fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num)
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.



```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

#Example

for power in powers_of_two(5):
    print(power)
```

Q7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(filename):
    try:
        file = open(filename, 'r')
        while True:
            line = file.readline()
            if not line:
                break
            yield line.strip()
    finally:
        file.close()


#Example

filename = 'example.txt'

for line in read_file_line_by_line(filename):
    print(line)
```

Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
tuples_list = [(1, 3), (4, 1), (2, 5), (3, 2)]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print(sorted_list)
```

Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius_temps = [0, 20, 30, 40, 100]
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print(fahrenheit_temps)
```

Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda x: x not in vowels, input_string))

input_string = "Hello World"
result = remove_vowels(input_string)
print(result)
```

Q11.  Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:image.png

```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.8],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python 3, Bernd Klein", 3, 24.99]
]

def calculate_total(orders):
    grand_total = 0
    for order in orders:
        order_number, book_info, quantity, price_per_item = order
        total_cost = quantity * price_per_item
        grand_total += total_cost
        print(f"Order Number: {order_number}")
        print(f"Book Title: {book_info}")
        print(f"Quantity: {quantity}")
        print(f"Price Per Item: {price_per_item}")
        print(f"Total Cost: {total_cost:.2f}")
        print("-" * 40)

    print(f"Grand Total: {grand_total:.2f}")

calculate_total(orders)
```
